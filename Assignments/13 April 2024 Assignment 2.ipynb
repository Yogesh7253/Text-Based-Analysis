```python
# Importing the regex library for regular expressions 

import regex as re
```

```python
# Defining a function named tokenize which returns the aplhabets using regex

def tokenize(text):
    return re.findall('[\w-]*[a-z|A-Z][\w-]*',text)
```

```python
# Calling our function tokenize

testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

```python
# Importing the Natural Language Toolkit library

import nltk
```

```python
# Storing the stopwords from stopwards package and printing them 

stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

```python
# Defining a function named remove_stop which returns other words except stopwords

def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]

cleantokens= remove_stop(testtokens)
print(cleantokens)
```

```python
# Converting the text in lowercase and storing it in pipeline

pipeline = [str.lower, tokenize, remove_stop]
```

```python
# Defining a function named prepare which inputs text and pipeline and returns tokens

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
# Making a new column tokens which stores the tokens

df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

```python
# Making a new column num_tokens which stores the length of tokens

df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

```python
# Importing Counter package from collections library

from collections import Counter
```

```python
# Testing our package

tokens = tokenize("She likes my cats and my cats like my sofa.")
testcount= Counter(tokens)
print(testcount)
```

```python
# Counting each token

counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
# we put the word frequency in it’s own dataframe for analysis

freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
# Describing the frequency 

freq_df.describe()
```
